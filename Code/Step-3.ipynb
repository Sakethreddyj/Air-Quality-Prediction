```python
# Calculating AQI for all Pollutants
# Function to Calculate AQI of PM2.5 Particulate Matter
def aqi_pm25(a):
    conc_pm = a
#     print(conc_pm)
    aqi_range = [0.0,50.0,100.0,200.0,300.0,400.0,500.0]
    conc_range = [0,30,60,90,120,250]

    if conc_pm > conc_range[0] and conc_pm <= conc_range[1]:
        return conc_range[0],conc_range[1],aqi_range[0],aqi_range[1]

    elif conc_pm > conc_range[1] and conc_pm <= conc_range[2]:
        return conc_range[1],conc_range[2],aqi_range[1],aqi_range[2]

    elif conc_pm > conc_range[2] and conc_pm <= conc_range[3]:
        return conc_range[2],conc_range[3],aqi_range[2],aqi_range[3]

    elif conc_pm > conc_range[3] and conc_pm <= conc_range[4]:
        return conc_range[3],conc_range[4],aqi_range[3],aqi_range[4]

    elif conc_pm > conc_range[4] and conc_pm <= conc_range[5]:
        return conc_range[4],conc_range[5],aqi_range[4],aqi_range[5]

    elif conc_pm > conc_range[5]:
          return conc_range[5],conc_pm,aqi_range[5],aqi_range[6]


def calculate_aqi_pm25(a):
    aqi_current = 0
    conc_current = a
#     print(conc_current)
    conc_low,conc_high,aqi_low,aqi_high = aqi_pm25(conc_current)
#     print(conc_low,conc_high,aqi_low,aqi_high)
    aqi_current = (( aqi_high - aqi_low ) * ( conc_current - conc_low ) / (conc_high - conc_low)) + (aqi_low)
#     print(aqi_current)
    return aqi_current
```

```python
df_imp['PM25_AQI'] = 0
for i in range(len(df_imp)):
    a = df_imp['PM25'].iloc[i]
    b = calculate_aqi_pm25(a)
    df_imp['PM25_AQI'].iloc[i] = b
```
